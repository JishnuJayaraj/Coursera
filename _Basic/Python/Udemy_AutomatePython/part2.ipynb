```python
### >>>>>>>>>>>>>>>>>>  STRINGS   <<<<<<<<<<<<< ###

sent1 = 'I love my cat'
sent2 = "I love jishnu's cat"   # getting a single quote inside
sent3 = 'she said \"I love jishnu\'s cat\" \n \t nice'   # use forward slash to escape character
sent4 = """ blah blah
chiku chiku boom boom
bhammmm!!!! """

sent4[10:17]
'boom' in sent4


# METHODS, they give a new string as STRINGS are immutable

spam = 'Hello world once again'
spam.lower()  # check islower()
spam.isdecimal()
spam.startswith('Hell')

'-'.join(['i','am','batman'])
'great-wall-of-china'.split('-')

'2391126'.ljust(10,'-')   #rjust: right justify, center()
'   empty'.strip()        # lstrip(), rstrip(), replace



## STRING FORMATTING

name = 'jishnu'
marks = 99
'this boy ' + name + ' has ' + str(marks) + ' points'
# conversion specifiers
'This boy %s has %s points' %(name, str(marks))





#### >>>>>>>>>>>>>>>>>>>>>>>>>>>>  REGULAR EXPRESSION   <<<<<<<<<<<<<<<<<<<<<<<<<<<<< ####
# specifying text patterns

import re
message = 'Jishnus (chullanX) number is 91-7736-136-692 and friends number is 9993-895-112'
message2 = 'batman has a batmobil which rides fast and batcopter to fly with lot of bats. cricket bats'

phnNumRegObj = re.compile(r'\d\d-\d\d\d\d-\d\d\d-\d\d\d')
matchObj=phnNumRegObj.search(message)            # or phnNumReg.findall()
matchObj.group()                                 # if it doesnt find, it returns None

# we can group using paranthesis
phnoRegObj2 = re.compile(r'(\d\d)-(\d\d\d\d-\d\d\d-\d\d\d)')
mo2 = phnoRegObj2.search(message)
mo2.group(1)
mo2.group(2)

# if we want to get the actual paranthesis, use escape character
phnoRegObj3 = re.compile(r'\(')
mo3 = phnoRegObj3.search(message)
mo3.group()

# if we want words with common prefix
batObj = re.compile(r'bat(man|mobil|copter|condom|cup)')
moBat = batObj.search(message2)
moBat.group(1)

# make a part optional, appear 0 or 1 time-> the part before '?' ::::: '*' for many times ::::: '+' 1 or more
# for exact number of times use '{3}'
phnoRegObj4 = re.compile(r'(\d\d\d\d-)?\d\d\d-\d\d\d')
mo4 = phnoRegObj4.search(message)
mo4.group()


# findall()
phnoRegObj5 = re.compile(r'\d\d\d\d-\d\d\d-\d\d\d')
mo5 = phnoRegObj5.findall(message)
mo5




#### >>> character class  <<<< ####
#  \d : digit 0-9                          \D : ! 0-9
#  \w : letter,digit,underscore            \W
#  \s : space,tab,newline                  \S


## making own character class, use []

# vowel class
vowelRegx = re.compile(r'[aeiouAEIOU]')  # r'[aeiouAEIOU]{3}  , 3 vowels consecutivly
vowelRegx.findall(message)

# negative character class
vowelRegx2 = re.compile(r'^[aeiouAEIOU]{4}')
vowelRegx2.findall(message)

# ^ start with ^
# $ end with   $
# . single character

message3 = 'First name: jishnu and Last name: jayaraj'
nameReg = re.compile(r'First name: (.*) Last name: (.*)')
nameReg.findall(message3)                        # greedy mode, add ? .. (.*?)

nameReg2 = re.compile(r'First name: \w+')    
nameReg2.findall(message3)
```
